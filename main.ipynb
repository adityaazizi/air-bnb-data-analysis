```python
def remove_outliers(col):
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1
    threshold = 1.5
    return col[(col >= (Q1 - threshold * IQR)) & (col <= (Q3 + threshold * IQR))]

numeric_columns = df.select_dtypes(include='number')
df_cleaned = numeric_columns.apply(remove_outliers)

non_numeric_columns = df.select_dtypes(exclude='number')
df = pd.concat([non_numeric_columns, df_cleaned], axis=1)
df.dropna(axis=0, inplace=True)

df
```

Analyze

```python
# Distribution Room Type
room_type_composition = df['room_type'].value_counts().reset_index()
room_type_composition.columns = ['room_type', 'count']
sns.barplot(data=room_type_composition, x='room_type', y='count')
```

```python
# Distribution City
city_composition = df['city'].value_counts().reset_index()
city_composition.columns = ['city', 'count']
sns.barplot(data=city_composition, x='count', y='city', orient='h')
plt.show()
```

```python
# Average Price by City
avg_price_by_city = df.groupby('city')['price'].mean().reset_index().sort_values(by='price', ascending=False)
sns.barplot(data=avg_price_by_city, x='price', y='city')
```
